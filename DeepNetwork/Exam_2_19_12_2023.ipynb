```python
first_weight = np.array([[-1, 1, 0],
                         [-1, 1, 0],
                         [-1, 1, 0]])
first_weight = np.stack((first_weight, first_weight, first_weight))

class CustomModel(nn.Module):
    def __init__(self):
        super(CustomModel, self).__init__()

        ### Asagisi
        self.convDown1 = nn.Conv2d(in_channels= 3, out_channels = 3, kernel_size = (9, 9), padding = (1, 1), stride = (2, 2))
        self.poolDown1 = nn.AvgPool2d(kernel_size= (6,6), padding = (1,1), stride = (3,3))


        ### Yukarisi

        self.convUp1 = nn.Conv2d(in_channels= 3, out_channels = 3, kernel_size = (3, 3), padding = (1, 1), stride = (1, 1))

        self.reluUp1 = nn.ReLU()
        self.maxPoolUp1 = nn.MaxPool2d(kernel_size= (2,2), padding = (0,0), stride = (2,2))

        self.convUp2 = nn.Conv2d(in_channels= 3, out_channels = 3, kernel_size = (8,8), padding = (1, 1), stride = (3, 3)) ##


       ## Ortak

        self.batch_norm = nn.BatchNorm1d(num_features = 169)
        self.sigmaoid = nn.Sigmoid()

        self.convShared1 = nn.Conv2d(in_channels= 3, out_channels = 3, kernel_size = (13, 13), padding = (1, 1), stride = (2, 2))
        self.convShared2 = nn.Conv2d(in_channels= 3, out_channels = 3, kernel_size = (5, 5), padding = (2, 2), stride = (2, 2))

        # YOUR IMPLEMENTATION GOES HERE



    def forward(self, image):

        ## Yukarisi

        print("***Upper Side***")

        conv1Up_out = self.convUp1(image)
        print('conv1Up_out', conv1Up_out.shape)

        reluUp1_out = self.reluUp1(conv1Up_out)
        print('reluUp1_out', reluUp1_out.shape)

        maxPoolUp1_out = self.maxPoolUp1(reluUp1_out)
        print('maxPoolUp1_out', maxPoolUp1_out.shape)

        convUp2_out = self.convUp2(maxPoolUp1_out)
        print('convUp2_out', convUp2_out.shape)

        print("***Upper Side over***")

       ## Asagisi

        print("***Lower Side***")

        convDown1_out = self.convDown1(image)
        print('convDown1_out', convDown1_out.shape)

        poolDown1_out = self.poolDown1(convDown1_out)
        print('poolDown1_out', poolDown1_out.shape)

        print("***Lower Side over***")


        ## Ortak

        print("***Combined***")

        combined1 = convUp2_out + poolDown1_out
        print('combined1', combined1.shape)

        batch_norm_out = self.batch_norm(combined1)
        print('batch_norm_out', batch_norm_out.shape)

        sigmoid_out = self.sigmaoid(batch_norm_out)
        print('sigmoid_out', sigmoid_out.shape)

        convShared1_out = self.convShared1(sigmoid_out)
        print('convShared1_out', convShared1_out.shape)

        convShared2_out = self.convShared2(convShared1_out)
        print('convShared2_out', convShared2_out.shape)

        # YOUR IMPLEMENTATION GOES HERE



        return image.permute(1, 2, 0)


# DO NOT MODIFY
# SHOWING YOUR RESULT

with torch.no_grad():
    model = CustomModel()
    modified = model(image)

    plt.figure()
    plt.imshow(modified)
```
